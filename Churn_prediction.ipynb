```python
def crossVal(x_train, y_train, model, fold =5 ):
    #x_train, y_train should be numpy arrary
    # modle is a test modle, e.g., #sklearn.linear_model.LinearRegression()
    kf = sklearn.cross_validation.KFold(len(x_train), fold) 
    mses = []
    for train_ind, test_ind in kf:
        train_x = x_train[train_ind]
        train_y = y_train[train_ind]
        test_x = x_train[test_ind] # for the current q, we only use x_train for KFold
        test_y = y_train[test_ind]
        
        md = model.fit(train_x, train_y)#sklearn.linear_model.LinearRegression().fit(train_x, train_y)
        y_hat = md.predict(test_x)
        mse = sklearn.metrics.mean_squared_error(test_y, y_hat)
        mses.append(mse)
    return np.mean(mses)

fullData['city'].unique()
df_city = pd.get_dummies(fullData['city'])
df_city.head()
df_phone = pd.get_dummies(fullData['phone'])
df_phone.head()
df_with_dummies = fullData
df_list = [df_phone,df_city,fullData]
df = pd.concat(df_list,axis =1)


df0 = fullData[fullData['churn'] ==1]
df1 = fullData[fullData['avg_rating_of_driver'].isnull()]
df2 = df1[df1.churn == 1]
print df0.shape[0]
print df1.shape[0]
print df2.shape[0]
print df2.shape[0]/float(df1.shape[0])

df3 = fullData[fullData['avg_rating_of_driver'] < 0.5*fullData['avg_rating_of_driver'].mean()]
df4 = df3[df3.churn == 1]

df5 = fullData[fullData['avg_rating_of_driver'] >= fullData['avg_rating_of_driver'].mean()]
df6 = df5[df5.churn == 1]

print df3.shape[0]
print df4.shape[0]
print df4.shape[0]/float(df3.shape[0])

print df5.shape[0]
print df6.shape[0]
print df6.shape[0]/float(df5.shape[0])
fullData.describe()
```
